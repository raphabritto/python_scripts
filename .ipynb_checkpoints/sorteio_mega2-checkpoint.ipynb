```python
import numpy as np
```

```python
apostas = []
min_numeros_aposta = 6
max_numeros_aposta = 15
min_mega = 1
max_mega = 60
```

```python
while True:
    try:
        numerosAposta = int(input('Quantidade de números por aposta: '))
        if numerosAposta < min_numeros_aposta or numerosAposta > max_numeros_aposta:
            raise ValueError('Digite um valor entre ' + str(min_numeros_aposta) + ' e ' + str(max_numeros_aposta) + '!')
    except ValueError as e:
        print(e)
        continue
    except:
        print('Erro desconhecido!')
        continue
    else:
        break
```
